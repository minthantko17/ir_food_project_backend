test cleaniing (by claude)

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
ps = PorterStemmer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def tokenize_and_stem(text):
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    tokens = [ps.stem(w) for w in tokens]
    return ' '.join(tokens)

sample = df.iloc[0]

# for search
ingredients = ' '.join(sample['RecipeIngredientParts'])
instructions = ' '.join(sample['RecipeInstructions'])
description = sample['Description'] if pd.notna(sample['Description']) else ''

combined = f"{sample['Name']} {description} {ingredients} {instructions}"

# test get image url
image_url = sample['Images'][0] if sample['Images'] is not None else None

print("=== Image URL (first only) ===")
print(image_url)

print("\n=== BEFORE cleaning ===")
print(combined[:300])

print("\n=== AFTER cleaning ===")
cleaned = clean_text(combined)
stemmed = tokenize_and_stem(cleaned)
print(stemmed[:300])
```
